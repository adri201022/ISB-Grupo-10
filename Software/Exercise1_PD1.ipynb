```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib widget
```

```python

N=300
n=np.arange(N)
n=n*2*np.pi/N
print(f"{n[0:6]}...{n[-7:-1]}")
```

```python
sin_data=np.sin(n)
cos_data=np.cos(n)
plt.figure(figsize=(6,6))
plt.plot(n,sin_data);plt.plot(n,cos_data)
```

```python
plt.figure(figsize=(6,6))
plt.plot(n,sin_data);plt.plot(n,cos_data,"r:")
plt.xlim([0,7]);plt.ylim([-1.4,1.4]);plt.legend(["Sin","Cos"])
plt.title("Sin and Cos functions");plt.xlabel("Time(s)");plt.ylabel("Function value")
plt.axhline(0,color="m")
```
